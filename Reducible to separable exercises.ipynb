### Exercise 1

$$\quad
\frac {dy}{dx} = (x+y+1)^2
$$
Change of variable :
$$\quad:
u=(x+y+1)
\quad \frac {du}{dx} = 1+\frac {dy}{dx}
$$
As:
$$\quad
 \frac {dy}{dx} = u^2
\quad u = tan (x+c)
$$
Finally:
$$\quad
y= tan(x+c)-x-1
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import math


def dy_dx(y, x):
    return (x+y+1)**2

xs = np.linspace(0,np.pi/5,100)
y0 = 0   # the initial condition
ys = odeint(dy_dx, y0, xs)
ys = np.array(ys).flatten()

plt.rcParams.update({'font.size': 14})  # increase the font size
plt.xlabel("x")
plt.ylabel("y")
plt.plot(xs, ys);
```

```python
print(xs)
y_exact_1 = np.tan(xs+np.pi/4)-xs-1 
y_exact_2 = np.tan(xs+2)-xs-1 
y_difference = ys - y_exact_1
plt.plot( xs, y_exact_1,xs,y_exact_2 ,"+");
```

```python
y_diff_1 = np.abs(y_exact_1 - ys)
y_diff_2 = np.abs(y_exact_2 - ys)
plt.semilogy(xs, y_diff_1)
plt.semilogy(xs, y_diff_2)
plt.ylabel("Error")
plt.xlabel("x")
```

### Exercise 2
$$
\quad
x(\frac {dy}{dx})=y(x^2*y^2-1)
$$
$$
\quad
u=xy \rightarrow y=\frac {u}{x}
$$
$$
\quad
du = y + x(\frac {dy}{dx})
$$
As:
$$
x(\frac {dy}{dx})=\frac {u}{x}(u^2-1)
$$
$$\quad
du= \frac{u^3}{x}
$$

Finally, solving the integral :
$$\quad
\int \frac{du}{u^3} = \int \frac{dx}{x}
$$
We find the analytical solution:
$$
\quad 
y=\frac{1}{x}* \sqrt{\frac{1}{2(k-ln|x|)}}
$$

```python
def dy_dx(y, x):
    return (y/x)*((x**2)*(y**2)-1)

xs = np.linspace(0,10,100)
y0 = 0  # the initial condition
ys = odeint(dy_dx, y0, xs)
ys = np.array(ys).flatten()

plt.rcParams.update({'font.size': 14})  # increase the font size
plt.xlabel("x")
plt.ylabel("y")
plt.plot(xs, ys);
```

```python
y_exact_1= (1/xs)*(np.sqrt(1/(2*(0.5-np.log(xs)))))
y_exact_2= (1/xs)*(np.sqrt(1/(2*(6.08-np.log(xs)))))
plt.plot( xs, y_exact_1,xs,y_exact_2,"+");
```

```python
y_diff_1 = np.abs(y_exact_1 - ys)
y_diff_2 = np.abs(y_exact_2 - ys)
plt.semilogy(xs, y_diff_1)
plt.semilogy(xs, y_diff_2)
plt.ylabel("Error")
plt.xlabel("x")
```
